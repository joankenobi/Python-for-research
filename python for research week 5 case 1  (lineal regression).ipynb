##5.1.3: Simple Linear Regression
In this video, you will...

>       Learn the basics of simple linear regression

```python
#debo repasar esto que es la base
def compute_rss(y_estimate, y):
  return sum(np.power(y-y_estimate, 2))
def estimate_y(x, b_0, b_1):
  return b_0 + b_1 * x
rss = compute_rss(estimate_y(x, beta_0, beta_1), y)
```

##5.1.4: Least Squares Estimation in Code
In this video, you will...

>     Learn how to use Python to compute the least squares estimate

>     Learn how to use Python to estimate parameter values that minimize the residual sum of squares (RSS) criterion

```python
rss=[]
slopes=np.arange(-10,15,0.001)
for slope in slopes:
  rss.append(np.sum((y-beta_0-slope * x)**2))
```

```python
rss
```

```python
ind_min=np.argmin(rss)
ind_min
```

```python
#en el video da 2.0
print(f"Estimate for the eslope: {slopes[ind_min]}")
```

```python
#plot figure
plt.figure()
plt.plot(slopes,rss)
plt.xlabel("slopes")
plt.ylabel("rss")
```

##5.1.5: Simple Linear Regression in Code
In this video, you will...

>     Learn how to fit a simple linear regression model in Python

```python
import statsmodels.api as sm
```

```python
n = 100
beta_0 = 5
beta_1 = 2
np.random.seed(1)
x = 10 * ss.uniform.rvs(size=n)
y = beta_0 + beta_1 * x + ss.norm.rvs(loc=0, scale = 1, size = n)
np.mean(x),np.mean(y)
```

```python
#creamos modelo
mod=sm.OLS(y,x)
mod
```

```python
#Estimate model
est=mod.fit()
est
```

```python
print(est.summary())
```

El modelo arriba nos da un valor estimado de $2.7569$ y sabemos de modelos anteriores que el valor que deberia dar es $2.0$.

Esto ocurre por que el modelo aplicado no considera las constantes y las intercepsiones, por lo que la pendiete parte del origen y no como la realizada con el modelo anterior.

```python
#Los slope son la pendiente y junto a las interceptions estas de encuentran el la funcionm asi que haciendo varias pruebas determiné la situación en la que el interceptor es negativo y el slope equivalente no lo considera.
n = 100
intercect1 = -5
slope1 = 9
np.random.seed(1)
x = 10 * ss.uniform.rvs(size=n)
y = intercect1 + slope1 * x + ss.norm.rvs(loc=0, scale = 1, size = n)
np.mean(x),np.mean(y)
```

```python
xx=np.array([0,10])
yy=intercect1+slope1*xx
yyy=8.5*xx

plt.plot(x,y,"o",ms=5)
plt.plot(xx,yy)
plt.plot(xx,yyy)
```

```python
#Mejorar el modelo para obtener una mejor pendiente
#creamos otro conjunto de variables x llamado X que considere las constantes y los elementos de su 1s

X=sm.add_constant(x)
```

```python
mod=sm.OLS(y,X)
est=mod.fit()
print(est.summary())
```

Para este caso la constante estimada es $5.2370$ y el valor estimado de $x1$ es de $1.9685$ que es mas cercano a $2.0$.

En si el valor de $5.2$ es la interception y el valos de $Y$ cuando todo predictor en el valor de $x1$ es $0$

El starnar error of the slope(pendiente) para ambos casos es $0.174$ y $0.031$.

##5.1.7: scikit-learn para regresión lineal

In this video, you will...

>      Learn how to generate data from a simple model

>      Learn how to fit a linear regression model using the scikit-learn library

>      Learn how to evaluate the fit of a model by determining the R**2 statistic 

```python
n=500
beta_0=5
beta_1=1
beta_2=-2
np.random.seed(1)
X_1=10*ss.uniform.rvs(size=n)
X_2=10*ss.uniform.rvs(size=n)
y=beta_0 + beta_1 * X_1 + beta_2 * X_2 + ss.norm.rvs(loc=0,scale=1,size=n)
```

```python
#apilamos nuetra entrada en una array (stack=apilar)
#los randoms arrays de X_1 y X_2 se acomodan en el nuevo array
X=np.stack([X_1,X_2], axis=1)
#Y=np.stack([X_1,X_2,y], axis=1)
```

```python
#muy util lo que hace
#X
```

```python
fig=plt.figure()
ax=fig.add_subplot(111, projection="3d")
ax.scatter(X[:,0],X[:,1],y,c=y)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_zlabel("$y$")
```

<h1>Creamos nuestro modelo de regresion lineal with "lm" 

```python
#supongo que aqui le decimos que si hay intercepcion
lm= LinearRegression(fit_intercept=True)
```

```python
lm.fit(X,y) #solo le pasamos la funcion de "y" and los datos de X que fueron creados aleatoriamentes

LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)
```

```python
#Encontramos los coheficientes y la intercepcion
print(f"Interception: {lm.intercept_}") #retorna el punto de interception
print(f"Coeficientes: {lm.coef_} beta_1={lm.coef_[0]} beta_2={lm.coef_[1]}") #retorna los coeficientes
```

<h1>Aplicamos la prediccion con nuestro modelo

```python
#We want a output for ours x_1 and x_2
X_0= np.array([2,4]) #ours x_1 and x_2
```

```python
#get a error
lm.predict(X_0)
```

```python
#for solve the error, because we have a single sample (como dice el error) we apply array.reshape(1,-1)
lm.predict(X_0.reshape(1,-1))
```

```python
# We can apply the ecuation for check our results
y_p= lm.intercept_+ lm.coef_[0]*2+ lm.coef_[1]*4
y_p
```

Greficar

```python
#debo aprender a graficar....
fig=plt.figure()
ax=fig.add_subplot(111, projection="3d")
ax.scatter(X[:,0],X[:,1],y,c=y)
ax.scatter(X_0[0],X_0[1],y_p,c="red", s=100)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_zlabel("$y$")
```
