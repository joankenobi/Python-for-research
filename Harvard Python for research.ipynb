Este problema es el más fácil, pero tenga cuidado de hacerlo bien.

Considere una perilla segura como la de la imagen de arriba. Tiene posiciones como números enteros de 0 a 99 (100 posiciones en total). Asumirá que siempre comienza en la posición 0. Si gira esta perilla en sentido antihorario, está aumentando la posición,
y si lo gira en el sentido de las agujas del reloj, está disminuyendo la posición. Si en algún momento alcanzas un giro completo, ya sea en el sentido de las agujas del reloj o en el sentido contrario, volverás a la posición 0.
Su misión en este ejercicio es escribir una función que reciba una lista de movimientos en esta caja fuerte y devuelva la posición final en la rueda después de aplicar todos los movimientos (asumiendo que comenzó en la posición 0). Cada movimiento será un número entero que será positivo si es un contra-
movimiento en el sentido de las agujas del reloj (aumentando la posición) y negativo si es un movimiento en el sentido de las agujas del reloj (disminuyendo la posición).

Ejemplo: si recibe dos movimientos, 10 y -5, comenzando desde 0, estará en la posición 10 después del primer movimiento en sentido antihorario, y luego estará en la posición 5 después del segundo que es en el sentido de las agujas del reloj. Entonces la respuesta
en este caso, será 5.


- cada movimiento será un número entero, por lo que puede ser desde el número entero mínimo hasta el número máximo máximo disponible en su idioma (incluido, por supuesto, 0)
- su función recibirá una lista de como máximo 10 mil movimientos.

```python
#Asi si funciona

class MyList(list):
  def remove_min(self):
    self.remove(min(self))
    return self
  def remove_max(self):
    self.remove(max(self))
    return self
    
the_list=list(range(0,8))

y=MyList(the_list)
print(the_list)

print(y.remove_max())
```

```python
class MyList(list):
    def remove_min(self):
        value = min(self)
        self.remove(value)
        return self

    def remove_max(self):
        value = max(self)
        self.remove(value)
        return self
        

u = [10, 3, 5, 1, 2, 7, 6, 4, 8]
i = MyList(u)

print(i.remove_max())
print(i.remove_min())
```
